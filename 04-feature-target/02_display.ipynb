```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# note: change this based on your setup
data_path = "./data/features_targets.csv"
ohlc_cols = ["Open", "High", "Low",	"Close"]
target_col = "TARGET_LOGRET_10"
```

```python
features_targets = pd.read_csv(data_path, index_col="Timestamp", parse_dates=["Timestamp"])
features_targets = features_targets.drop(ohlc_cols, axis=1)
features_targets
```

```python
# targets
targets = features_targets[[target_col]]
targets.describe()
```

```python
#features
features = features_targets.drop(target_col, axis=1)
features.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [30, 15] # default size
```

```python
targets.plot()
```

```python
features.plot()
```

```python
features.head(1000).filter(like="RSI").plot()
```

```python
# removing features with high corr to each other
corrs = features.corr()
sns.heatmap(corrs)
```

```python
# see if low variance features
features.plot.hist(subplots=True, legend=True, figsize=(15,50))
```
